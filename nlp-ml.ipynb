```python
from transformers import pipeline
generator = pipeline("text-generation", model="gpt2",framework="pt")
```

```python
prompt = "In a world where robots rule,"
result = generator(prompt, max_length=50)[0]['generated_text']
print(result)
```

```python
#overview-generative,pretrained for large dbs from internet,attention mechanism and context aware
```

```python
 #diff sizes and has limiations like biaseed info
```

```python
#exploration and analysis
prompt = "In the future, AI will"

for temp in [0.3, 0.7, 1.0, 1.3]:
    print(f"\n--- Temperature: {temp} ---")
    output = generator(prompt, max_length=60, temperature=temp, do_sample=True)[0]['generated_text']
    print(output)
```

```python
#analyzing styles,coherence and creativity
#lower temp-formal , makes sense and facts (less creative)
#for higher temp may go off topic
#lower temp-- we get the top prediction and med temp we get sampling from more possibilities
#high temp--> less likely or rare words
```

```python
# RESEARCH QUESTIONS
#-->when temp are tweaked the style , coherence and creativity gets affected
#it can summarize and continue and also produces biased sentences because its trained on internet texts
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Generate text with GPT-2
text = generator("AI is transforming the world of", max_length=100)[0]['generated_text']

# Create a WordCloud
wordcloud = WordCloud(width=800, height=400).generate(text)

# Display it
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
#ethical considerations
#it has bias beacause of its learning from internet and it can even generate wrong content , so we need to check before uplading or using the text from gpt-2
#so we need to consider prompt filtering and human moderation
```
